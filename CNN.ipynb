```python
import numpy as np



def CNN(input:np.ndarray,filter:np.ndarray):
    input_size = input.shape[0]
    filter_size = filter.shape[0]
    output_size = int((input_size - filter_size) + 1)
    output = np.zeros((output_size,output_size))  
    for i in range(output_size): 
        for j in range(output_size):  
            region = input[i:i+3, j:j+3]
            output[i, j] = np.sum(region * filter,dtype=np.int16)
            
    
            sum_values = []
            for row in range(3):
                for col in range(3):
                    sum_values.append(f"({region[row, col]} x {filter[row, col]})")

          
            sum_expression = " + ".join(sum_values)
            print(f"{sum_expression} = {output[i, j]}")
            print("\n")
            
    return output
    
      


matrix_R = np.array([
    [0,0,0,0,0,0],
    [0,156,155,156,158,158],
    [0,153,154,157,159,159],
    [0,149,151,155,158,159],
    [0,146,146,149,153,158],
    [0,145,143,143,148,158]])


filter_R = np.array([
    [-1,-1,1],
    [0,1,-1],
    [0, 1 , 1 ]   
    ])







output_R = CNN(matrix_R,filter_R)




```

```python
import numpy as np

def CNN(input: np.ndarray, filter: np.ndarray, stride: int = 1):
    input_size = input.shape[0]
    filter_size = filter.shape[0]
    output_size = int(((input_size - filter_size) / stride) + 1)
    output = np.zeros((output_size, output_size))
    
    for i in range(0, output_size * stride, stride):
        for j in range(0, output_size * stride, stride):
            region = input[i:i+filter_size, j:j+filter_size]
            output[i // stride, j // stride] = np.sum(region * filter, dtype=np.int16)
            
            sum_values = []
            for row in range(filter_size):
                for col in range(filter_size):
                    sum_values.append(f"({region[row, col]} x {filter[row, col]})")
            
            sum_expression = " + ".join(sum_values)
            print(f"{sum_expression} = {output[i // stride, j // stride]}")
            print("\n")
    
    return output

matrix_R = np.array([
    [0, 0, 0, 0, 0, 0],
    [0, 156, 155, 156, 158, 158],
    [0, 153, 154, 157, 159, 159],
    [0, 149, 151, 155, 158, 159],
    [0, 146, 146, 149, 153, 158],
    [0, 145, 143, 143, 148, 158]
])

filter_R = np.array([
    [-1, -1, 1],
    [0, 1, -1],
    [0, 1, 1]
])

stride_value = 1  # يمكن تغيير قيمة السترايد هنا
output_R = CNN(matrix_R, filter_R, stride=stride_value)
```
